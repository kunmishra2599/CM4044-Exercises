## Matplotlib Exercise

For this exercise, we are using Matplotlib to visualize Company Sales Data, which is stored in a text file, "company_sales_data.csv". It is available in NTULearn.

The data file has 9 columns of data with a header for every column's name:

"month_number","facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer","total_units","total_profit"

### 1.
Read total profit of all months from "company_sales_data.csv" and show it using a line plot:

Total profit data provided for each month. Generated line plot must include the following properties: â€“

- X label name = Month
- Y label name = Total profit
- figure title: Company profit per month
- x ticks: month number
- y ticks: 100000, 200000, 300000, 400000, 500000

Solution:

```python
## Solution 1
## use plt package
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt("company_sales_data.csv", delimiter = ",", skip_header = 1, dtype = int)
# print(data)
profits = data[:,8]
# print(profits)

%matplotlib inline    

x = data[:,0]
y = profits
plt.plot(x,y)

plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.xticks(data[:,0])
plt.title("Company profit per month")
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()

```

```python
## Solution 2
# use OOP (figure and axes) to draw the plot

fig = plt.figure(figsize=(4,3), dpi=100)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y, color="blue", linewidth=1.0, linestyle="-")
ax.set_xticks(data[:,0])
ax.set_yticks([100000, 200000, 300000, 400000, 500000])
ax.set_title("Company profit per month")
ax.set_xlabel("Month")
ax.set_ylabel("Total Profit")

plt.show()
```

### 7.

Use subplots to display sales data for Bathingsoap, Facewash, Facecream and Toothpaste.

Solution:

```python
## The best solution
# use OOP (figure and axes) to draw the plot

fig, ax = plt.subplots(2,2, figsize=(14,12))
ax = ax.flatten()

ax[0].plot(data[:,0],data[:,4], linestyle = "dotted", marker = "o", linewidth = 3, label = "bathingsoap")
ax[1].plot(data[:,0],data[:,1], linestyle = "dotted", marker = "o", linewidth = 3, label = "facecream")
ax[2].plot(data[:,0],data[:,2], linestyle = "dotted", marker = "o", linewidth = 3, label = "facewash")
ax[3].plot(data[:,0],data[:,3], linestyle = "dotted", marker = "o", linewidth = 3, label = "toothpaste")

titles  = ["bathingsoap", "facecream", "facewash", "toothpaste"]
for x in range(len(ax)):
    ax[x].set_xlabel("Month")
    ax[x].set_ylabel("Sales unit in number")
    ax[x].set_title(titles[x])

plt.show()
```

### 8.

Read all product sales data and show them using the stack plot.

- legend is on best position, so that it doesn't overlap with plot
- different colours are used for different product
- Figure Title: All product sales data in stack plot
- x title: Month
- y title: Sales units in number

Solution:

```python
## Solution 1
# use OOP (figure and axes) to draw the plot
fig = plt.figure(figsize=(12,7), dpi=80)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.stackplot(data[:,0],data[:,1],data[:,2],data[:,3],data[:,4],data[:,5],data[:,6], labels= ["facecream", "facewash", "toothpaste","bathingsoap","shampoo", "moisturiser"])

ax.set_xlabel("Month")
ax.set_xticks(data[:,0])
ax.set_xlim(1,12)
ax.set_ylabel("Sales unit in number")
ax.set_title("All product sales in stack plot")
ax.legend(loc="best", prop = {'size':10}, bbox_to_anchor=(0.17, 0.98) )

plt.show()
```
